```python
def remove_special_char(text):
    x=""
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+" "
    return x
```

```python
def stopwords_filter1(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words("english"):
            x.append(i)
    y=x[:]
    x.clear()
    return y
        
```

```python
def stopwords_filter2(text):
    stop=set(stopwords.words("english"))
    return [word for word in text.split() if word.lower() not in stop]
```

# Stemming 

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
y=[]
def stem_words(text):
    for i in text :
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
    #return [ps.stem(tok) for tok in text]
```

```python
stem_words(["loving","love","loved","walk","walking","moving"])
```

```python
df2["review"]=df2["review"].apply(stem_words)
df2.sample(1)
```

```python
def join_back(text):
    return " ".join(text)
```

```python
df2["review"]=df2["review"].apply(join_back)
df2.sample(1)
```
